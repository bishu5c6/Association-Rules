```python
dataset = [["Milk",'Onion','Nutmeg',"KidneyBeans",'Eggs',"yogurt"],
            ["Dill",'Onion','Nutmeg',"KidneyBeans",'Eggs',"yogurt"],
            ["Milk",'Apple',"KidneyBeans",'Eggs'],
            ["Milk",'Unicorn','Corn',"KidneyBeans","yogurt"],
            ["Corn",'Onion','Onion',"KidneyBeans",'Icecreams',"yogurt"]]
```

```python
import pandas as pd
from mlxtend.preprocessing import TransactionEncoder
```

```python
dataset
```

```python
te=TransactionEncoder()
te_ary = te.fit(dataset).transform (dataset)
```

```python
#converting into Dataframe
df = pd.DataFrame(te_ary, columns=te.columns_)
```

```python
df
```

```python
from mlxtend.frequent_patterns import apriori,association_rules
```

```python
frequent_itemsets = apriori(df, min_support=0.005, use_colnames = True)
```

```python
frequent_itemsets
```

```python
#Assocition Rule Mining
from mlxtend.frequent_patterns import association_rules
res = association_rules(frequent_itemsets,metric='confidence',min_threshold=0.7)
```

```python
res
```

```python
res1 = res[['antecedents','consequents','support','confidence','lift']]
res1
```

```python
res2=res1[res1['confidence']>=1]
res2
```
